```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coeff = np.array([
    -0.0073450325968958405,
    0.06534792286549167,
    -0.2561357051151653,
    0.5659344750939901,
    0.017230219148902437,
    1.0619550027665652,
    -0.6644047289080158,
    0.2787189053884093,
    -0.06890240706930176,
    0.0076013484260243295])
```

```python
buflen = 623
```

```python
ddelay = np.arange(buflen-len(coeff), buflen)
```

```python
f_sample = 48000
f_A = 440
f_tone = f_A*2**(-30/12)
div_ratio = f_sample / f_tone
```

```python
f_norm_test = np.arange(0.1, 20, 0.1)/div_ratio
```

```python
f_char = (np.exp(2j*np.pi*f_norm_test[:,None]*-ddelay[None, :])*coeff[None,:]).sum(axis=1)
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(np.arange(0.1, 20, 0.1), np.abs(f_char))
ax[1].plot(np.arange(0.1, 20, 0.1), np.angle(f_char))
ax[0].set_xticks(np.arange(0, 21))
ax[0].grid()
ax[1].set_xticks(np.arange(0, 21))
ax[1].grid()
```
